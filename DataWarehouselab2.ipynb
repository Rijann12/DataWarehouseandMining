```python
with open('/content/drive/MyDrive/DATALab/space (1).txt','r') as f:
  lines = f.readlines()[1:]
transactions = [line.strip().split(',')[1:] for line in lines]

for t in transactions[:50]:
  print(t)
```

```python
# One-hot encode the transactions(to convert data into binary format so algorlike apriori can understand)
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd

te = TransactionEncoder()
encoded_array = te.fit(transactions).transform(transactions)
df = pd.DataFrame(encoded_array, columns=te.columns_)

# Preview one-hot encoded DataFrame
df.head()
```

```python
# Apply Apriori Algorithm
from mlxtend.frequent_patterns import apriori, association_rules

frequent_itemsets_ap = apriori(df, min_support=0.2, use_colnames=True)
rules_ap = association_rules(frequent_itemsets_ap, metric="confidence", min_threshold=0.6)

print(" Frequent Itemsets (Apriori):\n", frequent_itemsets_ap)
print("\n Association Rules (Apriori):\n", rules_ap[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```

```python
#  Apply FP-Growth Algorithm
from mlxtend.frequent_patterns import fpgrowth

frequent_itemsets_fp = fpgrowth(df, min_support=0.2, use_colnames=True)
rules_fp = association_rules(frequent_itemsets_fp, metric="confidence", min_threshold=0.6)

print(" Frequent Itemsets (FP-Growth):\n", frequent_itemsets_fp)
print("\n Association Rules (FP-Growth):\n", rules_fp[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```

```python
# Compare Algorithms
comparison = {
    "Algorithm": ["Apriori", "FP-Growth"],
    "Candidate Generation": ["Yes", "No"],
    "Efficiency": ["Slower", "Faster"],
    "Memory Usage": ["High", "Low"],
    "Best Use Case": ["Smaller datasets", "Larger datasets"]
}

compare_df = pd.DataFrame(comparison)
print(" Algorithm Comparison:\n", compare_df)
```
